Programa para depurar los datos del detector y calcular únicamente los datos de interés, los cuales son: El contador de muones, el tiempo efectivo de medición y el flujo de muones.

```python
##############################################################################################
# Introduzca el nombre del archivo con los datos del detector en "datos_detector.txt",
# este archivo NO debe contener encabezados, solo los datos numéricos.
# El programa devolverá un nuevo archivo llamado "datos_detector_depurados.txt" con los resultados de interés:
# - Evento: número de muones registrados
# - Tiempo_efectivo: tiempo total de medición menos el tiempo muerto, en segundos
# - Flujo_muones: flujo de muones en s^{-1}, calculado como el número de muones dividido entre el tiempo efectivo
##############################################################################################


import numpy as np

# Cargar los datos desde el archivo de texto
# La matriz tiene dimensión Nx6
Matriz_datos = np.loadtxt("datos_detector.txt")

# Extraigo las columnas de la matriz
Evento = Matriz_datos[:, 0] 
Tiempo = Matriz_datos[:, 1] / 1000 # Extriago el tiempo en segundos (está en milisegundos en el archivo)
Voltaje_ADC = Matriz_datos[:, 2]
Voltaje_mV = Matriz_datos[:, 3]
Tiempo_muerto = Matriz_datos[:, 4] / 1000 # Extraigo el tiempo muerto en segundos (está en milisegundos en el archivo)
Temperatura = Matriz_datos[:, 5]

# Calculo el tiempo efectivo de medicion restando el tiempo muerto al tiempo total
Tiempo_efectivo = Tiempo - Tiempo_muerto

# Calculo el flujo de muones en s^{-1}
Flujo_muones = Evento / Tiempo_efectivo

# Construyo una matriz con los resultados de interés
Resultados = np.column_stack((Evento, Tiempo_efectivo, Flujo_muones))

# Guardo los resultados en un nuevo archivo de texto
np.savetxt("datos_detector_depurados.txt", Resultados, header="Evento Tiempo_efectivo_(s) Flujo_muones_(s^-1)", fmt = ["%d", "%.3f", "%.6f"], delimiter=" ")
```
