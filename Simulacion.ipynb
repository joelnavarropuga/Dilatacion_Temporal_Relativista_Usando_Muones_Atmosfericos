Programa para simular sesiones de medición con los detectores de muones CosmicWatch en las localizaciones que se ofrecen.

```python
##############################################################################################
# Al ejecutar el programa se le solicitará al usuario que elija un lugar para medir el flujo de muones atmosféricos, cada lugar tiene una altitud diferente, lo que afecta al flujo de muones que se mide.
# Se generará un archivo de texto con 1500 eventos, cada evento tendrá un tiempo efectivo (en segundos) y un flujo de muones (en s^-1) generado aleatoriamente siguiendo una distribución gaussiana centrada 
# en torno al valor esperado del flujo para la altitud del lugar elegido.
# El archivo se guardará con el nombre "Distribucion_flujo_muones_simulada.txt (puede cambiarlo si quiere)"
# ##############################################################################################


import math
import random
from datetime import datetime

locations = {
    1: ("Playa de Motril, 5.3 metros", 5.3),
    2: ("Facultad de Ciencias, Granada, 669 metros", 669),
    3: ("Villa de Otura, 810 metros", 810),
    4: ("Ermita Nueva de Dilar, 1032.3 metros", 1032.3),
    5: ("Cruz de la Atalaya, Padul, 1182.7 metros", 1182.7),
    6: ("El Dornajo, Sierra Nevada, 1691 metros", 1691),
    7: ("Pradollano, Sierra Nevada, 2107 metros", 2107),
    8: ("Hoya de la Mora, Sierra Nevada, 2511 metros", 2511),
    9: ("Pico Veleta, Sierra Nevada, 3091 metros", 3091),
}

print("Ingrese el numero del lugar en el que desea medir el flujo de muones atmosfericos:")
print()
for idx, (name, _) in locations.items():
    print(f"{idx}. {name}")
print()

n = int(input("Opcion: "))

if n not in locations:
    raise ValueError("Opcion no valida.")

location_name, h = locations[n]
phi = 0.1466 * math.exp(h / 3655.52)
sigma = 0.03 * phi
rng = random.Random()

# Genero el archivo donde voy a guardar los valores de flujo de muones aleatorios generados
with open("Distribucion_flujo_muones_simulada.txt", "w") as archivo:
    archivo.write("Evento Tiempo_efectivo_(s) Flujo_muones_(s^-1)\n")

    for i in range(1, 1501):
        flujo = rng.gauss(phi, sigma)  # Genero un valor de flujo siguiendo una distribución gaussiana centrada en phi y con anchura sigma
        t_eff = i / flujo
        archivo.write(f"{i} {t_eff:.4f} {flujo:.4f}\n")

print(f"Archivo generado para: {location_name}")
```
